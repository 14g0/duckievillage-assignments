```python
import numpy as np
import tensorflow as tf
import tensorflow.keras as keras
import tensorflow.keras.layers as layers

imgs = np.load("/tmp/images.npy", allow_pickle=True)
labels = np.load("/tmp/labels.npy", allow_pickle=True)

X = (np.asarray(imgs).astype('float32')/255.0)
y = np.asarray(labels)

p = 0.9
N = X.shape[0]
idx = np.random.permutation(N)
ntr = int(p*N + 0.5)
id_tr, id_te = idx[:ntr], idx[ntr:]
X_tr, X_te = X[id_tr], X[id_te]
y_tr, y_te = y[id_tr], y[id_te]

modelo = keras.Sequential([
    keras.Input(shape=X_tr.shape[1:]),
    layers.Conv2D(32, 3, strides=1, activation='relu'),
    layers.Conv2D(32, 3, strides=2, activation='relu'),
    layers.Conv2D(16, 3, strides=2, activation='relu'),
    layers.Conv2D(8,  3, strides=2, activation='relu'),
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(32, activation='relu'),
    layers.Dense(2)
])
modelo.compile(optimizer='Adam', loss='mse', metrics=['mse'])
modelo.summary()

hist = modelo.fit(
    X_tr, y_tr,
    validation_split=0.1,
    epochs=60,
    batch_size=128,
    verbose=0
)

y_pred = modelo.predict(X_te, verbose=0)
mse = np.mean((y_pred - y_te)**2)
print("MSE test:", mse)

modelo.save("./e2e_cnn.h5")
print("Modelo salvo em assignments/regression-cnn/_correto_cnn.h5")
```
